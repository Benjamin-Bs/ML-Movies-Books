```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('netflix_titles.csv');
```

```python
# Relevante Spalten
df = df[['title', 'release_year', 'duration']]
```

```python
# Entferne fehlende Werte
df = df.dropna()
```

```python
# Bereinigen der 'duration' Spalte (vereinfacht)
df['duration'] = df['duration'].str.extract('(\d+)')
df['duration'] = df['duration'].astype(float)
```

```python
# Text-Vektorisierung für den 'title' (Titel der Filme)
vectorizer = TfidfVectorizer(stop_words='english', max_features=100)
X_title = vectorizer.fit_transform(df['title']).toarray()
```

```python
# Unabhängige und abhängige Variablen definieren
# Kombiniere die Textdaten (X_title) mit den anderen numerischen Features (release_year)
X = np.hstack([X_title, df[['release_year']].values])  # Feature: 'title' (numerisch), 'release_year'
y = df['duration']  # Zielvariable: 'duration'
```

```python
# Daten aufteilen in Trainings- und Testset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Lineares Regressionsmodell erstellen und trainieren
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Vorhersage treffen
y_pred = model.predict(X_test)
```

```python

# Modellbewertung
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mittlerer quadratischer Fehler (MSE): {mse:.2f}")
print(f"Bestimmtheitsmaß (R^2): {r2:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', label="Vorhersagen vs. Tatsächliche Werte")
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label="Ideale Vorhersage")
plt.xlabel("Tatsächliche Dauer (Minuten)")
plt.ylabel("Vorhergesagte Dauer (Minuten)")
plt.title("Lineare Regression: Vorhersage der Dauer eines Films")
plt.legend()
plt.show()
```
